```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
PATH = PROJECT_ROOT / "data" / "sample" / "sample_daily.csv"
df = pd.read_csv(PATH)


# parse date robustly
df["date"] = pd.to_datetime(df["date"].astype(str).str.strip(), errors="coerce")

# assert no missing dates
if df["date"].isna().any():
    print("Unparsed dates:")
    print(df.loc[df["date"].isna(), "date"].head(20))

df = df.sort_values("date").reset_index(drop=True)
df.head()
```
